## A bit of Prep Work

```r
read.housing = function(file = 'kc_house_data.csv'){
  ## Read the csv file
  kc.housing <- read.csv(file, header = TRUE, 
                      stringsAsFactors = FALSE)
}

kc.housing = read.housing() 
```

```r
options(repr.plot.width=8, repr.plot.height=8)
require(car)
require(ggplot2)
```

## They Tend to be Cheaper & Better

```r
scatterplotMatrix(~ price + grade
                  | factor(floors)
                  , data = kc.housing)
```

The "rug plot" at the bottom of the price density shows a clustering of 3-story houses toward the left end of the price spectrum. Similarly, along the price v. grade scatter plot, you can see a clustering of 3-story houses higher grades and lower price.

```r
ggplot(kc.housing, aes(price)) + 
geom_density(aes(color = factor(floors)))
```

```r
ggplot(kc.housing, aes(price)) + 
geom_density(aes(color = factor(floors))) + 
xlim(0, 2000000) +
ggtitle('Price and Availability, Based on Floor Count')
```

Plenty of 1-3 story houses available below \$500K.  Massive spike of 3.5 story houses for a litter more than \$500K

```r
ggplot(kc.housing, aes(x = grade, y = price)) +
geom_point(aes(color = factor(floors)), alpha = 0.3)
```

```r
ggplot(kc.housing, aes(x = factor(grade), y = price, 
                       fill = floors))+
geom_violin(trim = TRUE, 
            draw_quantiles = c(0.25, 0.5, 0.75))
```

```r
ggplot(kc.housing, aes(x = grade, y = price)) +
stat_binhex(bins=10, aes(color = factor(floors)))
xlab('grade') + ylab('price') + ggtitle ('Price and Grade')
```

```r
ggplot(kc.housing, aes(x = grade, y = price))+ 
stat_binhex(bins=10, aes(color = factor(floors))) +
geom_density2d() + 
xlab('grade') + ylab('price') + ggtitle ('Price and Grade')
```

Countours suggest a clustering of grade 6 and grade 7 houses at 3.5 stories; 3-story seems a bit scattered.

```r
ggplot(kc.housing, aes(x = grade, y = price))+ 
stat_binhex(bins=10) + 
geom_density2d() + 
xlab('grade') + ylab('price') + ggtitle ('Price and Grade')
```

```r
ggplot(kc.housing, aes(x = grade, y = price)) +
stat_binhex(bins=10)
xlab('grade') + ylab('price') + ggtitle ('Price and Grade')
```

```r
ggplot(kc.housing, aes(x = factor(grade), y = price, 
                       fill = floors))+
geom_violin(trim = TRUE, 
            draw_quantiles = c(0.25, 0.5, 0.75))
```

```r
ggplot(kc.housing, aes(x = factor(floors), y = price, 
                       fill = grade))+
geom_violin(trim = TRUE, 
            draw_quantiles = c(0.25, 0.5, 0.75))
```

## They Tend to be Cheaper & Newer

```r
scatterplotMatrix(~ price + yr_built
                  | factor(floors)
                  , data = kc.housing)
```

```r
ggplot(kc.housing, aes(yr_built)) + 
geom_density(aes(color = factor(floors)))
```

```r
ggplot(kc.housing, aes(yr_renovated)) + 
xlim(2000, 2018) +
geom_density(aes(color = factor(floors)))
```

Massive influx of 3-story houses built after 2000; smaller spike in 3.5 story.
